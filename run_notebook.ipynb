```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import homography
```

### Load images in and subset to images of specific dimension (honeycomb example)

```python
honeycomb_img = np.float32(cv2.imread('dtd_folder\dtd\images\honeycombed\honeycombed_0003.jpg')) / 255.0
honeycomb_img = cv2.cvtColor(honeycomb_img, cv2.COLOR_BGR2RGB)
print(honeycomb_img.shape[:2])
plt.imshow(honeycomb_img)
plt.axis('off')
plt.title('Honeycomb Sample Image')
plt.show()
```

### Apply homography to warp normal input images

```python
H = homography.generate_homography_warp(honeycomb_img.shape[:2])
warped_img = homography.apply_homography(honeycomb_img, H)
```

```python
plt.imshow(warped_img)
plt.axis('off')
plt.title('Warped Honeycomb Sample Image')
plt.show()
```

### Apply homography to unwarp warped input images 

```python
vertices = homography.find_bounding_quadrilateral(warped_img)
```

```python
Hp = homography.generate_homography_unwarp(warped_img)
unwarped_img = homography.apply_homography(warped_img, Hp)
```
